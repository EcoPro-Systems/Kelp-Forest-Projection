```python
from plot_histogram_kelp import histogram_kelp
import matplotlib.pyplot as plt
import joblib
```

```python
file_path = "../Data/kelp_metrics_27_37.pkl"

# load data from disk
with open(file_path, 'rb') as f:
    data = joblib.load(file_path)

# plot time series
fig, ax, kdata = histogram_kelp(data)
plt.savefig(file_path.replace('.pkl', '_histogram_kelp.png'))
print(f"Saved histogram to {file_path.replace('.pkl', '_histogram_kelp.png')}")
plt.close()

# save data to pickle file
with open(file_path.replace('.pkl', '_histogram_kelp.pkl'), 'wb') as f:
    joblib.dump(kdata, f)
```
